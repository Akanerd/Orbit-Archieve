# **Indexing Array 2D Matriks**

```python
# Multi dimensional Arrays
# Numpy array memudahkan kita untuk membuat array multi dimensi
# Membentuk array 2 dimensi
# Fungsi shape array multi dimensi menghasilkan Tuple berisikan (Jumlah baris,Jumlah kolom)

import numpy as np
a = np.array([[0,1,2,3],
              [10,11,12,13]])
```

```python
# Shape
# Fungsi shape menhasilkan sebuah tuple yang berisikan panjang sebuah array pada tiap dimensi
a.shape
```
