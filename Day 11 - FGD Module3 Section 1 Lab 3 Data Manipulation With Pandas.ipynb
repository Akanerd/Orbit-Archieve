# **Handling Missing Values**

```python
import pandas as pd
import numpy as np
data1 = {'regiment': ['51st','29th', '2nd', '5th', '45th', '3rd',
'20th', '198th', '30th', '1st', '4th', '22nd'],
'trucks': ['MA7160', np.nan, 'BA3211', 'Tatra10', 'RT3121',
'Tatra10', 'WE2323','Tatra10', 'Tatra10', 'WE2323', 'WE2323',
'Tatra10'],
'tanks': ['Merkava Mark 4', 'Merkava Mark 4', 'Merkava Mark4', 'Leopard 2A6M', 'Arjun MBT', 'Arjun MBT', 'Merkava Mark 4', 'Arjun MBT', 'Leopard 2A6M', 'Arjun MBT', 'Leopard 2A6M', 'Arjun MBT'],
'aircraft':[ np.nan, 'none', 'none','Harbin Z-9', 'Harbin Z-9', 'Harbin Z-9', 'SH-60B Seahawk', 'SH-60B Seahawk', 'SH-60B Seahawk', 'SH-60B Seahawk','SH-60B Seahawk','SH-60B Seahawk' ]
}
df = pd.DataFrame(data1)
df
```

```python
df['location'] = np.nan #Create a new column with all NaN values
df
```

```python
df = df.dropna(axis=1, how='all') #Drop the column having NaN values
df
```

```python
#Drop the rows that contain less than 5 observations
#thresh itu jumlah value non NaN jadi pada intinya jika di baris tersebut jumlah data non NaN kurang dari thresh value maka baris tersebut dihilangkan
df.dropna(thresh=5)
df
```

# **Analysing Outliers**

```python
#We have a data stating heights of certain individuals and we try to detect outliers.
height_data = {'name':['Adit', 'Gia', 'Tim', 'Mark', 'Nur', 'Adil',
'Kia', 'Indar', 'Tia', 'Jack', 'Adele', 'Niall', 'Harry', 'Tom'],
'height': [5.9, 5.2, 5.1, 5.5, 4.9, 5.4, 6.2, 6.5, 7.1,
14.5, 6.1, 5.6, 1.2, 5.5]}
df = pd.DataFrame(height_data)
df
```

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']>max_threshold] #This is an outlier, above thanthe maximum limit
```

```python
df[df['height'] < min_threshold] #This is an outlier, below the minimum limit
```

```python
new_df = df[(df['height']<max_threshold) &
(df['height']>min_threshold)] #These are the values in between the two limits, i.e. without outliers
new_df
```
