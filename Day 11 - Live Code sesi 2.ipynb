# **Pie Chart**

```python
plt.figure(figsize=(8,8))
labels = 'forgs','hog','dog','Log'
fracs = [12,28,45,15]
explode = (0,0.05,0.1,0) # Jaarak antar suatu pie cghart
plt.pie(fracs,explode,labels=labels,autopct='%1.1f%%',shadow = True)
plt.show
```

# **Working with Text**

```python
mu, sigma = 100,15

x = mu+sigma*numpy.random.randn(10000)
n,bind,patches = plt.hist(x,50,density= True,facecolor='b',alpha=0.75)

plt.xlabel('Smart')
plt.ylabel('Probability')
plt.title('IQ Hitogram')
plt.text(60,.025,r'$\mu= 100,\ \sigma = 15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

# **Working with Multiple Figures and axes**

```python
t1 = np.arange(0.0,5.0,0.01)
t2 = np.arange(0.0,5.0,0.08)
f1 = np.exp(-t1)*np.cos(2*np.pi*t1)
f2 = np.exp(-t2)*np.cos(2*np.pi*t2)
plt.figure(figsize=(12,5))
plt.subplot(211)
plt.plot(t1,f1,'bo',t2,f2,'k') #nilai t1 terhadap f1
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```
