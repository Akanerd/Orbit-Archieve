# Set dalam Python

**Set** dalam bahasa pemrograman python adalah tipe data kolektif yang digunakan untuk menyimpan banyak nilai dalam satu variabel dengan ketentuan :
1.   Nilai anggota yang disimpan harus unik (tidak dupika)
2.   Nilai anggota yang sudah dimasukkan tidak bisa diubah lagi
3.   Set bersifat Unordered yakni tidak berurutan yang artinya tidak bisa diakses dengan index
**set** adalah salah satu 4 tipe data bawaan dalam python yang digunakan untuk menyimpan kumpulan data, 3 lainnya adalah daftar, Tuple dan dict semuanya dengan tujuan dan penggunaan yang berbeda.
**Set** merupakan bersifat tidak terurut danterindeks serta ditulis dengan kurung kurawal. Selain itu, set dengan sebuah list yang kita arahkan langsung ke dalam funsi set().





image.png

```python
# Menggunakan kurung kurawa 
himpunan_siswa = {'Andi','Lendis','Dicky','Wahid','Basith'}
print(himpunan_siswa)

# Mengkonversi list ke dalam set
himpunan_buah = set(['mangga','apel'])
print(himpunan_buah)

# Set dengan tipe data yang berbeda-beda
set_campuran = {'manusia','hewan',5,True,('a','b')}
print(set_campuran)

# Set bersifat Unordered alias tidak berurutan
```
