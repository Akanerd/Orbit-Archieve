# Program Konversi Bilangan Sederhana

*   Konversi bilangan desimal ke Biner
*   Konversi bilangan desimal ke Oktal
*   Konversi bilangan desimal ke Hexadesimal

Untuk membantu memahami bilangan biner, oktal dan hexadecimal
Program konversi bilangan desimal berbasis 10

```python
while True:
    number = int(input("Masukkan bilangan berbasis 10\n"))
    #pilih koversi
    print("a. Decimal menjadi Heksadesimal")
    print("b. Decimal menjadi Oktal")
    print("c. Decimal menjadi Biner")

    #Masukan Input
    print("Pilih konversi : ")
    choice = input()

    #variable pemilihan
    #kondisi
    #Heksadesimal dipilih jika "a"
    if choice == 'a':
      # lstrip helps remove "0x" from the left
      # rstrip helps remove "L" from the right
      # L represent a long number
          print("Heksadesimal dari bilangan " + str(number) +
        "adalah " + hex(number).lstrip("0x").rstrip("L"))
    elif choice == 'b' :  
      #Octal representation is done
      #by adding a prefix "0o"
          print("Heksadesimal dari bilangan " + str(number) +
        "adalah " + hex(number).lstrip("0o").rstrip("L"))
    elif choice == 'c' :
      #Binary represantation is done by
      #the addition of prefix "0b"
          print("Biner dari bilangan " + str(number) +
        "adalah "+ bin(number).lstrip("0b").rstrip("L"))
    else :
        print("Inputan anda salah")
        
    if input("Input lagi ? ")!= 'y':
            break
    

```
